# Definitions of variables

**UNITID** refers to Unit ID for institution

**INSTNM** refers to Institution name or college or university name

**CITY** refers to City that the college belongs to

**STABBR** refers to State postcode

**ZIP** refers to ZIP code for the college location

**ADM_RATE** refers to Admission rate i.e., the rate at which the students are being admitted

**UGDS** refers to Enrollment of undergraduate certificate/degree-seeking students

**TUITIONFEE_IN** refers to In-state tuition and fees i.e., the rate paid by students with permanent residence in the state in which their university is located

**MN_EARN_WNE_P10** refers to Mean earnings of students working and not enrolled 10 years after entry
